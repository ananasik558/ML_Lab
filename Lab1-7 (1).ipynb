2. Сбалансировано ли соотношение мужчин и женщин в наборе данных?

````python
if df[df['Gender'] == 'Male'].shape[0] == df[df['Gender'] == 'Female'].shape[0]:
    print("Сбалансировано")
else:
    print('Нет')
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
#пусть высокий уровень стресса это больше 50
low_df = df[df['Stress Levels'] <= 50]
high_df = df[df['Stress Levels'] > 50]

def convert_currency(val):
    new_val = val.replace('/', '')
    return float(new_val)

if low_df['Blood Pressure (s/d)'].apply(convert_currency).mean() >= high_df['Blood Pressure (s/d)'].apply(convert_currency).mean():
    print("Нет")
else:
    print("Да")
````

4. Правда ли, что уровень холестерина у мужчин и женщин отличается?

````python
men_df = df[df['Gender'] == 'Male']
women_df = df[df['Gender'] == 'Female']

if men_df['Cholesterol Level (mg/dL)'].mean() != women_df['Cholesterol Level (mg/dL)'].mean():
    print('Да')
else:
    print('Нет')
````

5. Правда ли, что с возрастом женщины имеют свойство оценивать свои когнитивные способности хуже, чем в юности (порог юности - 30 лет, женщина в возрасте - 60 лет+)

````python
young_df = df[(df['Gender'] == 'Female') & (df['Age (years)'] <= 30)]
old_df = df[(df['Gender'] == 'Female') & (df['Age (years)'] >= 60)]

if young_df['Cognitive Function'].mean() > old_df['Cognitive Function'].mean():
    print('Да')
else:
    print('Нет')
````

6. Влияет ли физическая активность на вес в возрасте более 50 лет?

````python
high_df = df[(df['Age (years)'] >= 50) & (df['Physical Activity Level'] == 'High')]
medium_df = df[(df['Age (years)'] >= 50) & (df['Physical Activity Level'] == 'Moderate')]
low_df = df[(df['Age (years)'] >= 50) & (df['Physical Activity Level'] == 'Low')]

if min(high_df['Weight (kg)'].mean(), medium_df['Weight (kg)'].mean(), low_df['Weight (kg)'].mean()) == high_df['Weight (kg)'].mean():
    print('Чем больше активность, тем меньше вес')
elif min(high_df['Weight (kg)'].mean(), medium_df['Weight (kg)'].mean(), low_df['Weight (kg)'].mean()) == medium_df['Weight (kg)'].mean():
    print('Люди со средней активностью весят меньше всех')
else:
    print('Неактивные люди весят меньше всех')
````

7. Сравните минимальный вес мужчин с высоким и низким заработком

````python
high_df = df[(df['Gender'] == 'Male') & (df['Income Level'] == 'High')]
low_df = df[(df['Gender'] == 'Male') & (df['Income Level'] == 'Low')]

if high_df['Weight (kg)'].mean() < low_df['Weight (kg)'].mean():
    print('Мужчины с высоким доходом весят меньше')
elif high_df['Weight (kg)'].mean() > low_df['Weight (kg)'].mean():
    print('Мужчины с низким доходом весят меньше')
else:
    print('Вес у мужчин с высоким и низким доходом примерно одинаков')
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
high_df = df[df['Income Level'] == 'High']
low_df = df[df['Income Level'] == 'Low']

def convert_currency(val):
    new_val = val.replace('High', '1').replace('Moderate', '0.5').replace('Low', '0')
    return float(new_val)

if high_df['Physical Activity Level'].apply(convert_currency).mean() > low_df['Physical Activity Level'].apply(convert_currency).mean():
    print('Да')
else:
    print('Нет')
````
